#### [ 'id',  'title', 'type',  'description', 'genres', 'production_countries', 'name']

```python
title = title[[ 'id',  'title', 'type',  'description', 'genres', 'production_countries', 'name']]
```

```python
title.shape
```

```python
title.isnull().sum()
```

```python
title.rename(columns = {'name':'cast'}, inplace = True)
```

```python
title.dtypes
```

```python
title.head()
```

```python
title['description'] = title['description'].apply(lambda x: str(x).split(' '))
```

```python
title.head()
```

```python
title['cast'] = title['cast'].apply(lambda x: str(x).split(','))
```

```python
title.head()
```

```python
def tag(text):
    li = []
    for i in text:
        if i not in ['[',']',"'"]:
            li.append(i)
    return li
            
```

```python
title['production_countries'] = title['production_countries'].apply(tag)
```

```python
title['production_countries'] = title['production_countries'].apply(lambda x:''.join(x))
```

```python
title['production_countries'] = title['production_countries'].apply(lambda x: str(x).split(','))
```

```python
title.head()
```

```python
title['type'] = title['type'].apply(lambda x: str(x).split(' '))
```

```python
title.head()
```

```python
title.dtypes
```

```python
title['genres'] = title['genres'].apply(tag)
title['genres'] = title['genres'].apply(lambda x:''.join(x))
title['genres'] = title['genres'].apply(lambda x: str(x).split(','))
```

```python
title['tags'] = title['type'] + title['description'] + title['production_countries'] + title['cast'] + title['genres']
```

```python
title = title[['id', 'title', 'tags']]
```

```python
title.dropna(inplace = True)
```

```python
title.isnull().sum()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text:
        y.append(ps.stem(i))
    return (' ').join(y)
        
```

```python
title['tags'] = title['tags'].apply(stem)
```

```python
title.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words = 'english', max_features = 5000)
```

```python
vector1 = tfidf.fit_transform(title['tags']).toarray()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words = 'english', max_features = 5000)
```

```python
vector = cv.fit_transform(title['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
import seaborn as sns
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
simi = cosine_similarity(vector1)
```

```python
similarity
```

```python
simi
```

```python
def recommend(movie):
    movie_index = title[title['title']==movie].index[0]
    sim = similarity[movie_index]
    movies_list = sorted(list(enumerate(sim)), reverse =True, key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(title.iloc[i[0]].title)
```

```python
def recommender(movie):
    movie_index = title[title['title']==movie].index[0]
    sim = simi[movie_index]
    movies_list = sorted(list(enumerate(sim)), reverse =True, key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(title.iloc[i[0]].title)
```

```python
recommend('Swades')
```

```python
recommender('Swades')
```
